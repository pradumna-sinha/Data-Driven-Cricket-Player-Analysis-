```python
# Importing the Libraries
import pandas as pd
import numpy as np
import json
```

```python
# Getting the data from batting_summary.json and putting it inside the a DataFrame
with open('D:/Final Year Project Cricket Analysis/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
```

```python
df_batting.head()
```

```python
# Adding a column to check if the player is out or not
# lambda function to check if dismissal has a value and putting out if it does
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 0 if len(x)>0 else 1)
df_batting.head()
```

```python
# Renaming the player names and removing the symbols
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
# function to change data type from object to numeric
def changetype(df):
    return pd.to_numeric(df,errors = 'coerce')
```

```python
df_batting['runs'] = changetype(df_batting['runs'])
df_batting['balls'] = changetype(df_batting['balls'])
df_batting['4s'] = changetype(df_batting['4s'])
df_batting['6s'] = changetype(df_batting['6s'])
df_batting['SR'] = changetype(df_batting['SR'])
```

```python
df_batting.info()
```

```python
# Finding the mean for all the players and grouping them into a df
#df = df_batting[['batsmanName','runs','balls','4s','6s','SR','out/not_out']]
df = df_batting.groupby('batsmanName').mean()
df.dropna(inplace = True)
df.head()
```

```python
# dividing the df into Train and Test sets
X = df.drop(columns = ['runs'])
y = df['runs']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,
                                                  random_state = 42,
                                                   test_size = 0.2,
                                                  shuffle = False)
```

```python
# Importing the Regressor models
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Putting the 1st 3 estimators
estimators = [
    ('rf', RandomForestRegressor(n_estimators=10, random_state=42)),
    ('lr', LinearRegression()),
    ('gbdt',GradientBoostingRegressor())
]
```

```python
from sklearn.ensemble import StackingRegressor

# Finally passing the individual predictor sets through a linear regressor
reg = StackingRegressor(
    estimators=estimators, 
    final_estimator=LinearRegression(),
    cv=10
)
```

```python
# Training the model on training data
reg.fit(X_train, y_train)
```

```python
y_train
```

```python
# After the model "reg" is trained we pass the total data to predict
y_pred = reg.predict(X)
```

```python
y_pred = pd.DataFrame(y_pred)
#d f_pred = pd.concat([X, y_pred], axis=1)
y_pred.reset_index(inplace = True)
```

```python
# Adding Index for primary key purposes
X.reset_index(inplace = True)
X.reset_index(inplace = True)
X.head()
```

```python
# Merging the dataframes into a single one
df_bat_pred = pd.merge(X,y_pred,how = 'inner',left_on = 'index',right_on = 'index')
```

```python
# Removing the index
df_bat_pred.drop(columns = ['index'],inplace = True)
# Renaming y_pred as Predicted Runs
df_bat_pred.rename(columns = {0:'Predicted Runs'},inplace = True)
df_bat_pred.head()
```

```python
from sklearn.metrics import r2_score

# check R^2 between x_true and y_pred
y_test_pred = reg.predict(X_test)
r2_score(y_test,y_test_pred)
```

```python
reg.fit(X_train, y_train)
```
